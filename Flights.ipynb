```python
import gzip
import csv
import json
import pandas as pd
import datetime
```

```python
#reading 1000 rows from data
#data = pd.read_csv('flightlist.csv.gz', nrows=1000, compression='gzip',error_bad_lines=False)

data = pd.read_csv('flightlist.csv.gz', compression='gzip', error_bad_lines=False)
data.head()
```

```python
#column list
list(data)
```

```python
#to drop if any value in the row has a nan
df_clean = data.dropna(how='any')
df_clean
```

```python
df_clean['firstseen'] = df_clean['firstseen'].str.strip('+00:00')
```

```python
df_clean
```

```python
df_clean['lastseen'] = df_clean['lastseen'].str.strip('+00:00')
```

```python
df_clean
```

```python
df_clean['day'] = df_clean['day'].str.strip('+00:00')
```

```python
df_clean
```

```python
#splitting firstseen and creating 2 new columns

df_split = df_clean['firstseen'].str.split(n=1, expand=True)
df_split.columns = ['date', 'time']
df_split
```

```python

df_clean['firstseen_date'] = df_split['date']
```

```python
df_clean['firstseen_time'] = df_split['time']
```

```python
df_clean
```

```python
#dropping the original firstseen column 
del df_clean['firstseen']
```

```python
df_clean
```

```python
#splitting lastseen and creating 2 new columns

df_split2 = df_clean['lastseen'].str.split(n=1, expand=True)
df_split2.columns = ['date', 'time']
df_split2
```

```python
df_clean['lastseen_date'] = df_split2['date']
```

```python
df_clean['lastseen_time'] = df_split2['time']
```

```python
df_clean
```

```python
#dropping the original lastseen column 
del df_clean['lastseen']
```
