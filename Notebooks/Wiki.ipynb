```python
import pandas as pd 
from sqlalchemy import create_engine
import string
```

```python
letters = string.ascii_uppercase
letters
```

```python
for letter in letters:
    url = 'https://en.wikipedia.org/wiki/List_of_airports_by_IATA_airport_code:_' + letter
    tables = pd.read_html(url)
    if letter == 'A':
        df= tables[0]
        df.columns = df.columns.droplevel([1])
    else:
        df_2 = tables[0]
        df_2.columns = df_2.columns.droplevel([1])
        df = pd.concat([df, df_2], sort= True)
```

```python
final_df = df[~df["IATA"].str.startswith("-")]
```

```python
del final_df['DST']
del final_df['Unnamed: 6_level_0']
del final_df['Unnamed: 7_level_0']
```

```python
final_df
```

```python
final_df.columns = [c.replace('\xa0', '_') for c in final_df.columns]
```

```python
final_df.rename (columns = {'Airport_name': 'airport_name'})
```

```python
final_df.columns = [a.lower() for a in final_df.columns]
final_df
```

```python
final_df.to_csv("..\Output\Airport_Codes.csv", index=False)
```
