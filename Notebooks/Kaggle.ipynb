```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
file = r"..\Data\airlines.csv"
airport_df = pd.read_csv(file, encoding = "UTF-8")
airport_df
```

```python
airport_df["Alias"] = airport_df["Alias"].replace(r"\N", np.NaN)
airport_df["ICAO"] = airport_df["ICAO"].replace(r"\N", np.NaN)
airport_df["Callsign"] = airport_df["Callsign"].replace(r"\N", np.NaN)
airport_df["Country"] = airport_df["Country"].replace(r"\N", np.NaN)
```

```python
airport_df
```

```python
airport_df["Alias"] = airport_df["Alias"].replace('-', np.NaN)
airport_df["ICAO"] = airport_df["ICAO"].replace('-', np.NaN)
airport_df["Callsign"] = airport_df["Callsign"].replace('-', np.NaN)
airport_df["Country"] = airport_df["Country"].replace('-', np.NaN)
```

```python
airport_df
```

```python
airport_df.drop(index=0, inplace= True)
airport_df.drop(index=1, inplace=True)
airport_df
```

```python

#find individual nulls
null_columns = airport_df.columns[airport_df.isnull().any()]
airport_df[null_columns].isnull().sum()
```

```python
len(airport_df)
```

```python
airport_df["Alias"] = airport_df["Alias"].replace(np.NaN, "")
airport_df["ICAO"] = airport_df["ICAO"].replace(np.NaN, "")
airport_df["Callsign"] = airport_df["Callsign"].replace(np.NaN, "")
airport_df["Country"] = airport_df["Country"].replace(np.NaN, "")
```

```python
airport_df = airport_df.fillna("")
airport_df
```

```python
final_airport_df = airport_df.rename(columns = {"Airline ID": "open_flights_id", 
                                               "Name": "airline_name", 
                                               "Alias": "airline_alias",
                                              "IATA": "iato_code",
                                              "ICAO": "icao_code",
                                              "Callsign": "airline_callsign",
                                              "Country": "country",
                                              "Active": "active"})
final_airport_df = final_airport_df.reset_index(drop = True)
final_airport_df
```

```python
final_airport_df.to_csv("..\Output\Open_Flights_Kaggle.csv", index=False)
```

```python
connection_string = "postgres:postgres@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
final_airport_df.to_sql(name='open_flights', con=engine, if_exists= 'append', index=False)
```
